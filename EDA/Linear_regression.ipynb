```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv(r"D:\Boston_house_price\data\Boston-house-price-data.csv")
df.head()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["MEDV"],bins=30,kde=True)
plt.title("Distribution of House Price (Medv)")
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(
    df.corr(),
    cmap="coolwarm",
    annot=True,        # show values
    fmt=".2f",         # 2 decimal places
    linewidths=0.5,
    square=True
)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
important_features = ["RM", "LSTAT", "PTRATIO"]

for col in important_features:
    plt.figure(figsize=(8,5))
    sns.scatterplot(x=df[col],y=df["MEDV"])
    plt.title(f"{col} vs MEDV")
    plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.title("outlier Detection")
plt.show()
```

## EDA Summary

- MEDV shows slight right skew.
- RM has strong positive correlation with MEDV.
- LSTAT has strong negative correlation with MEDV.
- Data has no missing values.
- Some features contain outliers but are acceptable for tree-based models.
